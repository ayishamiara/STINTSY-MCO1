## The Dataset

There are three datasets being used for this notebook. They are taken from [World Happiness Report](https://www.kaggle.com/datasets/unsdsn/world-happiness?select=2017.csv) taken from kaggle about a country’s happiness score from 2017, 2018, and 2019. The data collected from Gallup World Poll, detailing the happiness scores and rankings of different countries. 

Each row or observation represents a country spread across 12 variables for 2017. The following are the descriptions of each variable in the 2017 datasets:

- **`Country`**: Name of the Country 
- **`Happiness.Rank`**: Rank of the country based on the Happiness Score
- **`Happiness.Score`**: A metric measured in 2016 by asking the sampled people the question: "How would you rate your happiness on a scale of 0 to 10 where 10 is the happiest"
- **`Whisker.high`**: The upper whisker
- **`Whisker.low`**: The lower whisker
- **`Economy..GDP.per.Capita.`**: The extent to which GDP contributes to the calculation of the Happiness score
- **`Family`**:  The extent to which their Social Support contributes to the calculation of the Happiness score
- **`Health..Life.Expectancy.`**:  The extent to which their Health Life Expectancy contributes to the calculation of the Happiness score
- **`Freedom`**: The extent to which their Freedom to make life choices contributes to the calculation of the Happiness score
- **`Generosity`**: The extent to which their Generosity contributes to the calculation of the Happiness score
- **`Trust..Government.Corruption.`**: The extent to which Perception of Corruption contributes to the calculation of the Happiness score
- **`Dystopia.Residual`**: The metric  is the Dystopia Happiness Score(1.85) + the Residual value or the unexplained value for each country


Each row or observation represents a country spread across 9 variables for 2018 and 2019. The following are the descriptions of each variable in the 2018 and 2019 datasets:

- **`Overall rank`**: Rank of the country based on the Happiness Score
- **`Country or region`**: Name of the Country or region the country belongs to
- **`Score`**: A metric measured by asking  the  sampled people the question: “How would you rate your happiness on a scale of 1-10”
- **`GDP per capita`**: The extent to which GDP contributes to the calculation of the Happiness score
- **`Social Support`**: The extent to which their Social Support contributes to the calculation of the Happiness score
- **`Healthy life expectancy`**: The extent to which their Health Life Expectancy contributes to the calculation of the Happiness score
- **`Freedom to make life choices`**: The extent to which their Freedom to make life choices contributes to the calculation of the Happiness score
- **`Generosity`**: The extent to which their Generosity contributes to the calculation of the Happiness score
- **`Perceptions of corruption`**: The extent to which Perception of Corruption contributes to the calculation of the Happiness score

### Kmeans Clustering

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
```

The first thing the researchers did was to import the libraries above which will be used for clustering.

```python
happiness_17_18_19_grouped_copy = happiness_17_18_19_grouped.copy()
happiness_17_18_19_grouped_dropped = happiness_17_18_19_grouped.drop(['Country'], axis=1)
mms = MinMaxScaler()
mms.fit(happiness_17_18_19_grouped_dropped)
data_transformed = mms.transform(happiness_17_18_19_grouped_dropped)
```

A copy of the happiness_17_18_19_grouped dataframe was created and the **`Country`** column was dropped since the .fit can only be used in a dataframe that contains numerical values. The dataframe was transformed into a scaled version afterward. 

```python
Sum_of_squared_distances = []
K = range(1,15)
for k in K:
    km = KMeans(n_clusters=k)
    km = km.fit(data_transformed)
    Sum_of_squared_distances.append(km.inertia_)  
```

The formula above was used on order to compute for the sum of squared distances. This will be used in finding and selecting the optimal k that will be used for clustering. 

```python
plt.plot(K, Sum_of_squared_distances, 'bx-')
plt.xlabel('k')
plt.ylabel('Sum_of_squared_distances')
plt.title('Elbow Method For Optimal k')
plt.show()
```

The plot above shows the Elbow method for the optimal k. This was used by the researchers to sleect the best k value that will be used for clustering. We can observe above that the graph starts to bend at the 4th k which is why it was selected by the researchers as the k for clustering.

```python
km = KMeans(n_clusters=4, random_state=1)
km.fit(happiness_17_18_19_grouped_dropped)
```

The researchers then used the kmeans algorithm to create 4 clusters from the happiness_17_18_19_grouped_dropped dataframe.

## Statistical Inference

### How has the happiness score for each country changed over time?

The researchers would like to see how each happiness score for each country changed over time. To conduct the tests, the researchers using a **two-tailed one sample mean test** with a confidence level of 95% to compare the different years being tested. As python uses a default confidence level of 95%, we do not need to change it. The scipy module will also be used to help with making the statistical inference for the tests being done.

```python
from scipy import stats as st
```

```python
hs17 = happiness17_df['Happiness Score']
hs18 = happiness18_df['Happiness Score']
hs19 = happiness19_df['Happiness Score']
```

```python
hs17_mean = np.mean(hs17)
hs18_mean = np.mean(hs18)
hs19_mean = np.mean(hs19)

print(" 2017 Happiness Score Mean: {:.2f}\n".format(hs17_mean), "2018 Happiness Score Mean: {:.2f}\n".format(hs18_mean), "2019 Happiness Score Mean: {:.2f}\n".format(hs19_mean))
```

#### a. 2017 and 2018

The researchers would like to test to see if there is a significant diffence in the average happiness scores from 2018 with the ones from 2017. Below are the Null and Alternative Hypotheses.

**Null Hypothesis: There is no difference in the average happiness scores data sets.**

**Alternative Hypothesis: There is a difference in the average happiness scores of the data sets.**

```python
st.ttest_1samp(hs18, hs17_mean)
```

From the result, the t-score is 0.2928 while the p-value is 0.7701. Following a significance level of 0.05, we accept the null hypothesis as the value of the p-value is bigger than the significance level. 

Therefore, there is no significant difference in the average happiness scores in 2018 and the average happiness scores in 2017 at a significance level of 5%.

#### b. 2018 and 2019

Following the previous test conducted on the years 2017 and 2018, the researchers would like to test to see if there is a significant diffence in the average happiness scores from 2019 with the ones from 2018. Below are the Null and Alternative Hypotheses.

**Null Hypothesis: There is no difference in the average happiness scores data sets.**

**Alternative Hypothesis: There is a difference in the average happiness scores of the data sets.**

```python
st.ttest_1samp(hs19, hs18_mean)
```

From the result, the t-score is 0.4267 while the p-value is 0.6702. Following a significance level of 0.05, we accept the null hypothesis as the value of the p-value is bigger than the significance level. 

Therefore, there is no significant difference in the average happiness scores in 2019 and the average happiness scores in 2018 at a significance level of 5%.

#### c. 2017 and 2019

Following the previous test conducted on the years 2018 and 2019, the researchers would like to test to see if there is a significant diffence in the average happiness scores from 2019 with the ones from 2017. Below are the Null and Alternative Hypotheses.

**Null Hypothesis: There is no difference in the average happiness scores data sets.**

**Alternative Hypothesis: There is a difference in the average happiness scores of the data sets.**

```python
st.ttest_1samp(hs19, hs17_mean)
```

From the result, the t-score is 0.7211 while the p-value is 0.4720. Following a significance level of 0.05, we reject the null hypothesis as the value of the p-value is bigger than the significance level. 

Therefore, there is no significant difference in the average happiness scores in 2019 and the average happiness scores in 2017 at a significance level of 5%.

* **`Cluster 0`**: Countries with Moderately High Happiness Scores
* **`Cluster 3`**: Countries with Moderately Low Happiness Scores
* **`Cluster 2`**: Countries with Low Happiness Scores

Generally, Happiness Score has a strong positive relationship with Life Expectancy and Economy, a moderate positive relationship with Freedom, and a weak positive relationship with Trust. However, looking at the countries by their cluster, some attributes play a more indicative role than others in inferring how high or low a country’s happiness is. Specifically, high Life Expectancy, Economy, and Freedom are potential indicators of a high Happiness Score. In contrast to this, low Trust and Economy are prospective basis for inferring a low Happiness Score. 

Lastly, the statistical inference shows no significant difference between the happiness scores of the countries for each of the three years. Hence, it has not made any significant changes throughout 2017 to 2019.
